### Introduzione ed equilibrazione

Il sistema oggetto di studio è lo stesso analizzato nell'<font face="Courier"> Esercitazione 4</font>, ossia un modello Lennard-Jones di Argon ($\sigma=0.34 nm$, $\epsilon/b_B=120 K$, $m=39.948 amu$) nelle seguenti condizioni:
* fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ ($r_c = 2.2$);
* fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ ($r_c = 2.5$);
* fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ ($r_c = 5.0$).

In questo caso effettuiamo le simulazioni in condizioni di *ensamble canonico* (numero di particelle $N$, volume $V$ e **T costante**) attraverso un algoritmo di Metropolis. Dunque, anzichè produrre le configurazioni risolvendo le equazioni del moto a $E$ costante (come viene fatto nell'<font face="Courier"> Esercitazione 4</font> in condizioni di *ensamble microcanonico*), in questo caso le configurazioni vengono generate a temperatura costante campionando una catena di Markov adeguata a realizzare il campionamento del peso statistico di Boltzmann.
Il passo dell'algoritmo $x_i \rightarrow x_i + \delta$ viene generato secondo una probabilità di transizione $T(\vec{x}|\vec{y})$ uniforme e, come ogni qualvolta abbiamo fatto uso di un algoritmo di Metropolis, la prima accortezza a cui è necessario porre attenzione è determinare la lunghezza del passo più consona a rispettare la regola empirica del $50\%$.
Dopo qualche tentativo ci si è accorti che per lo stato solido e liquido questo limite è facilmente raggiungibile adattando $\delta$, mentre per quanto riguarda lo stato gassoso le migliori accettazioni raggiunte sono circa del $62\%$. Il gas infatti possiede una densità molto più bassa delle altre fasi ed è quindi statisticamente facile riuscire a modificare la posizione delle particelle. Alla conclusione dei tentativi per ciascuna fase vengono selezionati i passi seguenti, con i quali vengono da qui in avanti svolte le simulazioni:
* $\delta_{solido}=0.11$;
* $\delta_{liquido}=0.2$;
* $\delta_{gas}=5.0$;

### Autocorrelazione e taglia dei blocchi

Attraverso un set di misure istantanee della simulazione vogliamo ora studiare l'andamento dell'autocorrelazione delle due grandezze termodinamiche di interesse in funzione dello step temporale.
Lo scopo di questa operazione è quello di mostrare i tempi di decadimento della correlazione degli elementi appartenenti alla catena di Markov che viene usata per il campionamento. Fatto questo, saremo in grado di dare una prima stima approssimativa della taglia minima dei blocchi in cui suddividere le misure (*blocking method*) per poter applicare a tutti gli effetti il CLT ed evitare quindi una sottostima dell'incertezza statistica dei risultati finali. 

---
L'autocorrelazione ci indica il grado di dipendenza tra elementi successivi della catena di Markov ed è calcolabile attraverso la formula seguente:

$$AC_\hat{O}(t)=\frac{1}{\sigma^2_\hat{O}}\Bigg[\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}\hat{O}(t')\hat{O}(t'+t)-\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}\hat{O}(t')\cdot\frac{1}{t_{max}-t}\sum_{t'=0}^{t_{max}-t}\hat{O}(t'+t) \Bigg] \quad \text{dove}\quad \sigma^2_\hat{O}=\frac{1}{t_{max}}\sum_{t'=0}^{t_{max}}\hat{O}^2(t')-\bigg(\frac{1}{t_{max}}\sum_{t'=0}^{t_{max}}\hat{O}(t')\bigg)^2$$

Qui di seguito vengono mostrati i grafici relativi all'autocorrelazione di pressione e energia potenziale per particella in funzione del tempo nelle tre diverse fasi del sistema in esame. Vengono usate $5\cdot10^5$ misure e un numero di passi che varia a seconda della fase di interesse. 

```python
def AutoCorrelation(n_points,obs):
    t=np.arange(n_points)
    t_max=len(obs)
    factor = [np.zeros(n_points) for j in range(3)]
    factor[0] = [sum(obs[0:t_max-i]*obs[i:t_max]) for i in t]/(t_max-t)
    factor[1] = [sum(obs[0:t_max-i]) for i in t]/(t_max-t)
    factor[2] = [sum(obs[i:t_max]) for i in t]/(t_max-t)
    factor4 = (np.std(obs))**2
    return (factor[0]-factor[1]*factor[2])/factor4   
```

```python
#Fase solida:
autocorrelationE = AutoCorrelation(250,solid_pot)
autocorrelationP = AutoCorrelation(250,solid_pressure)

plt.subplots(figsize=(13,5))
plt.subplot(1,2,1)
plt.title('Potential energy per particle',fontsize=16)
plt.plot(autocorrelationE,lw=3,color='#117733')
plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$AC(t)$',fontsize=14)
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Pressure',fontsize=16)
plt.plot(autocorrelationP,lw=3,color='#661100')
plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$AC(t)$',fontsize=14)
plt.grid(True)

plt.suptitle('Autocorrelation function: solid phase',fontsize=18)
plt.tight_layout()
plt.show()
####################################################################################
#Fase liquida:
autocorrelationE = AutoCorrelation(500,liquid_pot)
autocorrelationP = AutoCorrelation(500,liquid_pressure)

plt.subplots(figsize=(13,5))
plt.subplot(1,2,1)
plt.title('Potential energy per particle',fontsize=16)
plt.plot(autocorrelationE,lw=3,color='#117733')
plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$AC(t)$',fontsize=14)
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Pressure',fontsize=16)
plt.plot(autocorrelationP,lw=3,color='#661100')
plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$AC(t)$',fontsize=14)
plt.grid(True)

plt.suptitle('Autocorrelation function: liquid phase',fontsize=18)
plt.tight_layout()
plt.show()
###################################################################################
#Fase gassosa:
autocorrelationE = AutoCorrelation(250,gas_pot)
autocorrelationP = AutoCorrelation(250,gas_pressure)

plt.subplots(figsize=(13,5))
plt.subplot(1,2,1)
plt.title('Potential energy per particle',fontsize=16)
plt.plot(autocorrelationE,lw=3,color='#117733')
plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$AC(t)$',fontsize=14)
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Pressure',fontsize=16)
plt.plot(autocorrelationP,lw=3,color='#661100')
plt.xlabel(r'$t$',fontsize=14)
plt.ylabel(r'$AC(t)$',fontsize=14)
plt.grid(True)

plt.suptitle('Autocorrelation function: gas phase',fontsize=18)
plt.tight_layout()
plt.show()
```

Come possiamo osservare dai grafici, l'autocorrelazione per tutti i casi in esame descresce in maniera esponenziale.
Dopo circa $100$ passi per la fase solida le configurazioni campionate diventano sostanzialmente scorrelate. 
Dopo circa $350$ step vale lo stesso discorso per la fase liquida.
Per quanto riguarda la fase gassosa notiamo invece che l'autocorrelazione decade più velocemente rispetto alle altre fasi. Infatti è sufficiente un tempo pari a $t\sim 25$ per portare a zero l'autocorrelazione. 

Per un corretto utilizzo del *data blocking* la taglia dei blocchi deve essere tale da garantire che la correlazione sia limitata ai bordi del blocco. In questo modo le medie di blocco saranno scorrelate tra loro e potremo ricorrere al teorema del limite centrale.
Questo risultato rappresenta il limite inferiore al numero di dati che dobbiamo usare.

---
A questo punto dobbiamo chiederci quanti dati vale la pena inserire nei blocchi per avere una precisione statistica ottimale senza oltrepassare il "limite superiore", oltre il quale i costi computazionali aumenterebbero senza ragione.
Per far questo, tramite uno script python, studiamo l'andamento dell'incertezza statistica dei risultati finali ottenuti tramite *data blocking* di $U/N$ e di $P$ al crescere della dimensione dei blocchi $L$.
Il numero di misure utilizzate $M$ a tal fine rimane sempre lo stesso ad eccezione dei casi in cui la taglia dei blocchi $L$ non è un divisore esatto del totale delle misure.
In questo caso viene usato un numero di blocchi $N$ pari al numero intero risultato del rapporto tra $M$ e $L$. Questa approssimazione è del tutto trascurabile ai fini della nostra analisi.

Per la simulazione vengono usate $M=5\cdot10^5$ misure e la taglia dei blocchi $L$ và da $L=10$ a $L=5000$ a passi di $25$. Qui di seguito vengono mostrati i risultati.

```python
def error_blocking_method(L,obs):  #(L,N_blocks,obs)
    block_average=0
    block_squared_average=0
    N_blocks = int(len(obs)/L)
    for i in range(N_blocks):
        single_measure = sum(obs[i*L:(i+1)*L])/L
        block_average += single_measure
        block_squared_average += single_measure*single_measure
    block_average /= N_blocks
    block_squared_average /= N_blocks
    error = np.sqrt((block_squared_average-block_average**2)/(N_blocks-1))
    return error

#Block_size= [ i for i in range(10, 5000+1, 1) if(500000%i==0)]
#N_blocks = [int(500000/i) for i in Block_size]

Block_size =  [ i for i in range(10, 5000+1, 25)]
Labels ={'solid_pot':solid_pot,'solid_pressure':solid_pressure, 'liquid_pot':liquid_pot, 
            'liquid_pressure':liquid_pressure, 'gas_pot':gas_pot, 'gas_pressure':gas_pressure}
for label in Labels:
    sigma_measures = np.zeros(len(Block_size))
    index=0
    for l in Block_size:
        sigma_measures[index] = error_blocking_method(l,Labels[label])   #(l,N_blocks[index],Labels[label])
        index = index+1
    np.savetxt("error_"+label+".dat", sigma_measures)
    #print(label +' done.')
```

```python
#Fase solida:
plt.subplots(figsize=(13,5))
plt.subplot(1,2,1)
plt.scatter(Block_size, error_solid_pot, s=25, color='#117733', alpha=0.5)
plt.plot(Block_size, error_solid_pot, lw=1, color='#117733', alpha=0.5)
plt.title('Error U/N',fontsize=16)
plt.xlabel('Block size',fontsize=14)
plt.ylabel(r'$\sigma_{U/N}\,[J]$',fontsize=14)
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Error P',fontsize=16)
plt.scatter(Block_size, error_solid_pressure, s=25, color='#661100', alpha=0.5)
plt.plot(Block_size, error_solid_pressure, lw=1, color='#661100', alpha=0.5)
plt.xlabel('Block size',fontsize=14)
plt.ylabel(r'$\sigma_{P}\,[N/m^2]$',fontsize=14)
plt.grid(True)

plt.ticklabel_format(axis='both', style='sci',scilimits=(0,5))
plt.suptitle('Error trend using blocking method: solid phase',fontsize=18)
plt.tight_layout()
plt.show()
#####################################################################################
#Fase liquida:
plt.subplots(figsize=(13,5))
plt.subplot(1,2,1)
plt.scatter(Block_size, error_liquid_pot, s=25, color='#117733', alpha=0.5)
plt.plot(Block_size, error_liquid_pot, lw=1, color='#117733', alpha=0.5)
plt.title('Error U/N',fontsize=16)
plt.xlabel('Block size',fontsize=14)
plt.ylabel(r'$\sigma_{U/N}\,[J]$',fontsize=14)
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Error P',fontsize=16)
plt.scatter(Block_size, error_liquid_pressure, s=25, color='#661100', alpha=0.5)
plt.plot(Block_size, error_liquid_pressure, lw=1, color='#661100', alpha=0.5)
plt.xlabel('Block size',fontsize=14)
plt.ylabel(r'$\sigma_{P}\,[N/m^2]$',fontsize=14)
plt.grid(True)
plt.ticklabel_format(axis='both', style='sci',scilimits=(0,5))

plt.suptitle('Error trend using blocking method: liquid phase',fontsize=18)
plt.tight_layout()
plt.show()
#####################################################################################
#Fase gassosa:
plt.subplots(figsize=(13,5))
plt.subplot(1,2,1)
plt.scatter(Block_size, error_gas_pot, s=25, color='#117733', alpha=0.5)
plt.plot(Block_size, error_gas_pot, lw=1, color='#117733', alpha=0.5)
plt.title('Error U/N',fontsize=16,pad=14)
plt.xlabel('Block size',fontsize=14)
plt.ylabel(r'$\sigma_{U/N}\,[J]$',fontsize=14)
plt.ticklabel_format(axis='both', style='sci',scilimits=(-5,5))
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Error P',fontsize=16,pad=14)
plt.scatter(Block_size, error_gas_pressure, s=25, color='#661100', alpha=0.5)
plt.plot(Block_size, error_gas_pressure, lw=1, color='#661100', alpha=0.5)
plt.xlabel('Block size',fontsize=14)
plt.ylabel(r'$\sigma_{P}\,[N/m^2]$',fontsize=14)
plt.ticklabel_format(axis='both', style='sci',scilimits=(2,5))
plt.grid(True)

plt.suptitle('Error trend using blocking method: gas phase',fontsize=18)
plt.tight_layout()
plt.show()
```

Come possiamo notare dai grafici, l'incertezza statistica per taglie piccole è sottostimata. Dunque le medie di blocco, che costituiscono l'unità effettiva sulla quale si basa la statistica, sono tra loro correlate e perciò non sono soddisfatte le ipotesi del teorema del limite centrale. 
Aumentando il numero di dati per blocco, quello che succede è che le misure correlate vengono "spinte" ai bordi del singolo blocco e una volta che la taglia è sufficientemente grande questo meccanismo ci garantisce un'ottima indipendenza statistica tra le medie di blocco.
D'ora in avanti possiamo utilizzare il CLT e, di conseguenza, la stima dell'errore statistico come deviazione standard della media delle medie di blocco è a tutti gli effetti affidabile. 
Come possiamo vedere, dopo una certa taglia, l'errore tende a stabilizzarsi e quindi, un ulteriore avanzamento non produrrebbe alcun vantaggio (naturalmente a parità del numero totale di misure).
E' quindi questo il punto in cui possiamo fermarci e considerare la taglia del sistema sufficiente per i nostri scopi.

### Confronto tra le g(r) prodotte dagli algoritmi NVT e NVE

Contemporaneamente alle simulazioni effettuate per le stime di $U/N$ e $P$ vengono stampati i risultati finali della media a blocchi per la funzione di distribuzione radiale $g(r)$. Questa funzione esprime la probabilità di trovare una qualunque coppia di particelle a distanza $r$ confrontata con la medesima probabilità che si avrebbe nel caso di  particelle indipendenti (e quindi di un gas ideale). 
L'espressione usata per il calcolo di questa distribuzione è la seguente:

$$g(r)=\frac{1}{\rho N \Delta V(r)}\Bigg\langle \sum_j^N \sum_{i\neq j}^N \delta\big(r-|\vec{r}_i-\vec{r}_j|\big) \Bigg\rangle \quad\text{dove}\quad \Delta V(r) = \frac{4}{3} \pi \big[(r + dr)^3 − r^3 \big]$$

Da un punto di vista algoritmico la forma della $g(r)$ viene ottenuta suddividendo l'intervallo delle possibili distanze $[0,L_{box}/2]$ (in p.b.c.) tra due particelle all'interno del box di simulazione in $100$ sottointervalli
e, ad ogni step della simulazione, per ogni coppia che si trova alla distanza $r\in[r_i,r_{i+1}]$ vengono aggiunte due unità al bin $i-$esimo corrispondente.
Infine utilizzando il metodo a blocchi e normalizzando per la quantità $\frac{1}{\rho N \Delta V(r)}$ si ottiene una stima della $g(r)$ con relativa incertezza. 

Questa operazione viene eseguita sia nel codice NVT Monte Carlo sia nell'algoritmo di dinamica molecolare proposto nella <font face="Courier"> Esercitazione 4</font>. Per quanto riguarda quest'ultimo, per tutte e tre le fasi viene fatto ripartire il codice un numero di volte sufficiente ad ottenere una buona equilibrazione del sistema.
