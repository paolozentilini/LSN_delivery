```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
M=10000          
N=100                

unif_sampl = np.loadtxt("uniform_sampling.dat")
imp_sampl = np.loadtxt("importance_sampling.dat")
```

```python
x= np.arange(N,M,N)

plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.title("Uniform sampling (M = 1e4, N=100).")
plt.errorbar(x, unif_sampl.T[0], yerr=unif_sampl.T[1], color='red', alpha=0.5, label='Uniform', linewidth=2)
plt.hlines(1,0,10000, color='green', linestyle='--', label="Theoric expected value", linewidth=2, alpha=0.5)
plt.ylabel("Integral's value")
plt.xlabel("#throws")
plt.legend()
plt.grid(True)
plt.ylim(0.992,1.005)

plt.subplot(1,2,2)
plt.title("Importance sampling (M = 1e4, N=100).")
plt.errorbar(x, imp_sampl.T[0], yerr=imp_sampl.T[1], color='blue', alpha=0.5, label='Importance', linewidth=2)
plt.hlines(1,0,10000, color='green', linestyle='--', label="Theoric expected value", linewidth=2, alpha=0.5)
plt.ylabel("<Integral's value>")
plt.xlabel("#throws")
plt.legend()
plt.grid(True)
plt.ylim(0.992,1.005)

plt.show()
```

```python
drw= np.loadtxt("discrete_random_walk.dat")
crw= np.loadtxt("continous_random_walk.dat")
x=np.arange(0,100,1)
y=np.arange(0,100,0.1)
```

```python
plt.figure(figsize=(18,6))

plt.subplot(1,2,1)
plt.title("Discrete Random Walk (M = 1e4, N=100)")
plt.errorbar(x, drw.T[0], yerr=drw.T[1], color='blue', alpha=0.7, label='Simulation',marker='.',markersize=8, linestyle='None')
plt.plot(y,np.sqrt(y),color='green',label='sqrt(N)',linestyle='-',linewidth=1)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.xlabel(r'$N_{passi}$')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.title("Continous Random Walk (M = 1e4, N=100)")
plt.errorbar(x, crw.T[0], yerr=crw.T[1], color='red', alpha=0.7, label='Simulation',marker='.',markersize=8, linestyle='None')
plt.plot(y,np.sqrt(y),color='green',label= r'$k\sqrt{N}$',linestyle='-',linewidth=1)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.xlabel(r'$N_{passi}$')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(16,6))

plt.title("Random Walk (M = 1e4, N=100).")
plt.errorbar(x, crw.T[0], yerr=crw.T[1], color='red', alpha=0.7, label='Countinous case',marker='.',markersize=8, linestyle='None')
plt.plot(y,np.sqrt(y),color='green',label='sqrt(N)',linestyle='--',linewidth=1)
plt.ylabel("<Distance value>")
plt.xlabel("#steps")
plt.legend()
plt.grid(True)
plt.show()
```
