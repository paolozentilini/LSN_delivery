### Esercizio 3.1





```python
M=10000          
N=100                

call = np.loadtxt("continous_call.dat")
call1 = np.loadtxt("discrete_call.dat")
put = np.loadtxt("continous_put.dat")
put1 = np.loadtxt("discrete_put.dat")

x = np.arange(N,M,N)
```

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.title('Vanilla European call option price')
plt.errorbar(x, call.T[0], yerr=call.T[1], color='grey', label='Monte Carlo evaluation')
plt.hlines(14.975790778311286, 0, M, color='red', linewidth=1.5, label='B&S analytic solution', linestyle='--')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Vanilla European put option price')
plt.errorbar(x, put.T[0], yerr=put.T[1], color='grey', label='Monte Carlo evaluation')
plt.hlines(5.4595325819072364, 0, M, color='red', linewidth=1.5, label='B&S analytic solution', linestyle='--')
plt.legend()
plt.grid(True)

plt.suptitle('Direct Asset price evaluation', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.title('Vanilla European call option price',fontsize=14)
plt.errorbar(x, call1.T[0], yerr=call1.T[1], color='black', alpha=0.5, label='Monte Carlo evaluation')
plt.hlines(14.975, 0, M, color='red', linewidth=1.5, label='B&S analytic solution', linestyle='--')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.title('Vanilla European put option price',fontsize=14)
plt.errorbar(x, put1.T[0], yerr=put1.T[1], color='black', alpha=0.5, label='Monte Carlo evaluation')
plt.hlines(5.46, 0, M, color='red', linewidth=1.5, label='B&S analytic solution', linestyle='--')
plt.legend()
plt.grid(True)

plt.suptitle('Discrete Asset price evaluation', fontsize=20)
plt.show()
```
