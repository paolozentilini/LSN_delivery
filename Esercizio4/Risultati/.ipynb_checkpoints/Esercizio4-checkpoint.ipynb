### Introduzione ed equilibrazione del sistema

In questa esercitazione viene studiato un sistema di $108$ particelle disposte all'interno di un box con condizioni periodiche al contorno (p.b.c.) e interagenti entro la distanza di cut-off ($r_c$) mediante un potenziale di tipo Lennard-Jones della forma:

$$V(r)=4\epsilon\Bigg[\bigg(\frac{\sigma}{r}\bigg)^{12}-\bigg(\frac{\sigma}{r}\bigg)^6\Bigg].$$

Possiamo notare come l'energia potenziale, in questo caso, oltre ad avere una forma complicata, è una funzione delle posizioni delle molecole del sistema. Per questo motivo non esiste una soluzione analitica alle equazioni del moto e dunque l'unico modo per arrivare ad un risultato è integrarle in modo numerico.
A tal fine utilizziamo l'*algoritmo di Verlet* che, per il calcolo delle posizioni delle particelle ad un istante di tempo successivo, necessita non solo delle posizioni attuali, ma anche di quelle allo step antecedente: 
 
$$\vec{r}(t+\delta t)=2\vec{r}(t)-\vec{r}(t-\delta t) + \vec{a}(t)\delta t^2$$

con $\vec{r}(t)$ vettore rappresentativo della configurazione spaziale di ciascuna delle molecole all'istante di tempo $t$ e $\vec{a}(t)$ vettore delle accelerazioni.
Osserviamo come l'*algoritmo di Verlet* non faccia uso delle velocità delle particelle per determinare la configurazione successiva. Tuttavia, in seguito sarà comunque necessario il calcolo di queste ultime.

---

Il programma originale permette di simulare le condizioni di un sistema in *ensamble microcanonico*, ossia a volume $V$, numero di particelle $N$ ed energia $E$ costanti, partendo dalla configurazione iniziale <font face="Courier">config.0</font>.
Il codice genera poi casualmente le velocità iniziali richiedendo che il centro di massa del sistema non possieda moti di deriva.
Infine vengono riscalate le velocità affinchè l'energia cinetica si avvicini ad un valore compatibile con la temperatura target $\tilde{T}$ (alla quale si vogliono studiare le proprietà termodinamiche del sistema), per poi costruire la configurazione finale salvata in <font face="Courier">config.final</font>. Il fattore di riscalamento $f_{scale}$ viene ricavato calcolando innanzitutto $T(t+\delta t/2)$ nel modo seguente

$$T(t+\delta t/2)=\sum_{j=0}^{N_{part}-1}\big|\vec{v}_j\big(t+\delta t/2\big)\big|^2$$

per poi usare il rapporto tra questa temperatura e la temperatura target:

$$f_{scale}= \sqrt{\frac{\tilde{T}}{T(t+\delta t/2)}}$$

Il modulo quadro della velocità nella prima espressione viene calcolato eseguendo una media nel corso di un certo numero di passi (specificato in ciascuna simulazione) dell'algoritmo affinchè il suo valore risulti più preciso. Infatti il rischio è che il valore istantaneo delle velocità possa essere il risultato di una fluttuazione statistica e non della reale situazione cinetica della fase in considerazione.

---

Il nostro compito è inizialmente quello di abilitare l'algoritmo alla ripartenza dalla configurazione raggiunta nel suo predecente utilizzo in modo tale da poter equilibrare il sistema. Infatti, poiché la configurazione iniziale è prodotta artificialmente, il sistema necessita di una fase di assestamento più o meno lunga, al termine della quale viene raggiunto un equilibrio caratterizzato da una temperatura $T$ in generale diversa da quella target $\tilde{T}$.
La possibilità di ripartire più volte consente di risolvere questo problema.

Inoltre, per accelerare questa procedura, abbiamo inserito nel codice la possibilità di riscalare le velocità anche in fase di riavvio.
A questo scopo per ogni ripartenza viene eseguito inizialmente uno step dell'algoritmo e poi calcolata la velocità $\vec{v}(t+\delta t/2)$ in questo modo:

$$\vec{v}(t+\delta t/2)= \frac{\vec{r}(t+\delta t)-\vec{r}(t)}{\delta t}$$

Questa viene quindi riscalata ($\vec{v}\rightarrow \vec{v}_s$) per avvicinare la temperatura $T$ al valore target e con la nuova velocità viene aggiornata la configurazione $\vec{r}_{new}(t) = \vec{r}(t+\delta t) - \vec{v}_s \delta t$. 
Possiamo quindi calcolare a partire dalle configurazioni $\vec{r}(t+\delta t)$ e $\vec{r}(t-\delta t)$ la velocità

$$\vec{v}(t)= \frac{\vec{r}(t+\delta t)-\vec{r}(t-\delta t)}{2 \delta t}$$

con la quale misurare una delle grandezze termodinamiche di interesse, ossia l'energia cinetica $E_{kin}$.
 
**Osservazione**: 
Onde evitare di far ripartire il programma più volte, viene creata una funzione in grado di effettuare l'equilibrazione con un solo avvio dell'algoritmo.


Ripetendo questo procedimento siamo in grado di vedere come il sistema tende ad equilibrarsi verso la temperatura target $\tilde{T}$ di input contenuta in <font face="Courier">input.dat</font> come mostrato nel grafico che segue.

```python
P0=np.loadtxt("../../Esercizio4/Equilibrazione/output_epot.dat")
K0=np.loadtxt("../../Esercizio4/Equilibrazione/output_ekin.dat")
E0=np.loadtxt("../../Esercizio4/Equilibrazione/output_etot.dat")
T0=np.loadtxt("../../Esercizio4/Equilibrazione/output_temp.dat")

plt.figure(figsize=(16,10))
plt.title("System equilibration",fontsize=20,pad=15)
plt.axvspan(xmin=0, xmax=2000, color='grey',alpha=0.2,label='Equilibration phase')
plt.hlines(0.8,0,5000,color='black',ls='--',lw=1.5,label='Target temperature')
plt.plot(P0,label='Potential energy',color='#332288',lw=3,alpha=0.8)
plt.plot(K0,label='Kinetic energy',color='#999933',lw=3,alpha=0.9)
plt.plot(E0,label='Total energy',color='#CC6677',lw=3,alpha=0.9)
plt.plot(T0,label='Temperature',color='#661100',lw=3,alpha=0.8)
plt.ylabel("Measured quantities [L-J units]",fontsize=16)
plt.xlabel("# throws",fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
```

Notiamo come in questo caso, partendo inizialmente dalla configurazione solida *fcc*, il sistema impieghi circa 3-4 ripartenze per raggiungere un valore della temperatura $T$ prossimo a quello target.
Inoltre è importante sottolineare come l'energia cinetica (e quindi anche $T$) sia caratterizzata da un crollo deciso all'inizio della simulazione. Questo fatto non deve sorprenderci poiché il sistema, partendo da una configurazione cristallina perfetta con un'energia potenziale $E_{pot}$ molto bassa, tende ad uno stato più disordinato e dunque la condizione iniziale non è più recuperabile nel corso della dinamica.
Per questo motivo se $E_{pot}$ tende inizialmente ad aumentare, dato che ci troviamo in condizioni di *ensamble microcanonico*, l'energia totale viene conservata e questo porta $E_{kin}$ ad avere un crollo repentino alla partenza.

**Nota bene**: nella simulazione le grandezze termodinamiche sono misurate in unità di Lennard-Jones (unità ridotte). Questa scelta ha due vantaggi: in primis ci consente di esprimere tutte le grandezze in termini di massa, lunghezza ed energia e in secondo luogo, tutte le quantità misurate sono di ordine 1. In questo modo si evita il rischio che i calcoli standard in virgola mobile portino ad un risultato in over/underflow.
D'ora in avanti le grandezze di interesse vengono espresse in unità SI e quindi, vengono usati i seguenti riscalamenti:
$$ E =\epsilon E^*\quad,\quad T=\frac{\epsilon}{k_b}T^*$$
dove con l'asterisco vengono indicate le grandezze in unità ridotte.

---
A questo punto siamo pronti per simulare le tre fasi dell'Argon ricordando che per ciascuna di esse viene eseguita una fase di equilibrazione preliminare. Tutte le misure delle grandezze termodinamiche del sistema e le relative incertezze sono calcolate mediante il metodo *data blocking*.

### Solido:

Per la simulazione della fase solida vengono inizialmente eseguiti $1000$ step di equilibrazione durante i quali ogni $100$ passi vengono riscalate le velocità. Nei $10000$ step successivi il sistema viene lasciato evolvere senza riscalamenti e alla conclusione di questo processo vengono effettuate le misurazioni.

```python
T0=np.loadtxt("../../Esercizio4/Risultati/Solido/solid_equilibration_temp.dat")
T0=120*T0

plt.figure(figsize=(15,4))
plt.title("System equilibration",fontsize=16)
plt.plot(T0,color='#117733',lw=3,label='System temperature')
plt.axvspan(xmin=0, xmax=1000, color='grey',alpha=0.15,label='Equilibration phase')
plt.hlines(0.8*120,0,10850,color='black',ls='--',label='Target temperature')
plt.ylabel("Temperature [k]",fontsize=14)
plt.xlabel("N step",fontsize=14)
plt.grid(True)
plt.xlim(-300,11000)
plt.legend(fontsize=14)
plt.show()
```

Come possiamo vedere dal grafico i passi indicati sono sufficienti per arrivare ad una buona equilibrazione. Infatti, dopo una breve fase di regolazione, la temperatura si mantiene oscillante intorno al valore target di $96 K$. Ora siamo pronti a effettuare le misurazioni. 

```python
K=eps*np.loadtxt("../../Esercizio4/Risultati/Solido/solid_ekin.dat")
V=eps*np.loadtxt("../../Esercizio4/Risultati/Solido/solid_epot.dat")
E=eps*np.loadtxt("../../Esercizio4/Risultati/Solido/solid_etot.dat")
T=120*np.loadtxt("../../Esercizio4/Risultati/Solido/solid_temp.dat")

x = np.arange(1,len(K)+1,1)

plt.figure(figsize=(18,12))
plt.title(" ")
plt.subplot(2,2,1)
plt.errorbar(x,V.T[0], yerr=V.T[1], color='#332288', alpha=0.8, label='Potential energy', linewidth=2)
plt.ylabel("Potential energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,2)
plt.errorbar(x,K.T[0], yerr=K.T[1], color='#661100', alpha=0.8, label='Kinetic energy', linewidth=2)
plt.ylabel("Kinetic energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,3)
plt.errorbar(x,E.T[0], yerr=E.T[1], color='#882255', alpha=0.8, label='Total energy', linewidth=2)
plt.ylabel("Total energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,4)
plt.errorbar(x,T.T[0], yerr=T.T[1], color='#117733', alpha=0.8, label='Temperature', linewidth=2)
plt.ylabel("Temperature [K]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.suptitle('System measurements', fontsize=22, y=0.94)
plt.show()
```

I dati riportati nei quattro grafici sono relativi a una simulazione di $5\cdot10^4$ step suddivisi in $100$ blocchi per poter utilizzare efficacemente la tecnica del *data blocking*. 
L'andamento delle grandezze così misurate evidenzia un comportamento sufficientemente stabile e questa può essere la conseguenza di un'equilibrazione effettuata in maniera corretta.

### Liquido:

Nei 10000 step successivi il sistema viene lasciato evolvere senza riscalamenti e alla conclusione di questo processo vengono effettuate le misurazioni.

```python
T0=np.loadtxt("../../Esercizio4/Risultati/Liquido/liquid_equilibration_temp.dat")
T0=120*T0

plt.figure(figsize=(15,4))
plt.title("System equilibration",fontsize=16)
plt.plot(T0,color='#117733',lw=3,label='System temperature')
plt.axvspan(xmin=0, xmax=1000, color='grey',alpha=0.15,label='Equilibration phase')
plt.hlines(1.1*120,0,10750,color='black',ls='--',label='Target temperature')
plt.ylabel("Temperature [k]",fontsize=14)
plt.xlabel("N step",fontsize=14)
plt.grid(True)
plt.xlim(-300,11000)
plt.legend(fontsize=14)
plt.show()
```

```python
K=eps*np.loadtxt("../../Esercizio4/Risultati/Liquido/liquid_ekin.dat")
V=eps*np.loadtxt("../../Esercizio4/Risultati/Liquido/liquid_epot.dat")
E=eps*np.loadtxt("../../Esercizio4/Risultati/Liquido/liquid_etot.dat")
T=120*np.loadtxt("../../Esercizio4/Risultati/Liquido/liquid_temp.dat")

x = np.arange(1,len(K)+1,1)

plt.figure(figsize=(18,12))
plt.title(" ")
plt.subplot(2,2,1)
plt.errorbar(x,V.T[0], yerr=V.T[1], color='#332288', alpha=0.8, label='Potential energy', linewidth=2)
plt.ylabel("Potential energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,2)
plt.errorbar(x,K.T[0], yerr=K.T[1], color='#661100', alpha=0.8, label='Kinetic energy', linewidth=2)
plt.ylabel("Kinetic energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,3)
plt.errorbar(x,E.T[0], yerr=E.T[1], color='#882255', alpha=0.8, label='Total energy', linewidth=2)
plt.ylabel("Total energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,4)
plt.errorbar(x,T.T[0], yerr=T.T[1], color='#117733', alpha=0.8, label='Temperature', linewidth=2)
plt.ylabel("Temperature [K]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.suptitle('System measurements', fontsize=22, y=0.94)
plt.show()
```

L'andamento delle grandezze misurate evidenzia un comportamento meno stabile rispetto alla fase solida. Questo risultato non ci deve sorprendere essendo il liquido una fase più disordinata del gas e quindi meno semplice da equilibrare. Una tecnica che aiuta a ridurre le fluttuazioni è quella di eseguire simulazioni più lunghe aumentando il numero di step per blocco. Ciò sarà particolarmente utile al fine di studiare la fase gassosa.

### Fase gassosa:

Per la simulazione della fase gassosa vengono inizialmente eseguiti $2\cdot10^4$ step di equilibrazione durante i quali ogni 100 passi vengono riscalate le velocità.
Date le difficoltà nell'effettuare una buona fase di equilibrazione, oltre ai passaggi del codice descritti nell'introduzione, viene aggiunto uno strumento ulteriore
Partendo da una temperatura iniziale di $120 K$, ogni $2000$ step viene aggiunta la quantità $+2.4 K$ fino a raggiungere il valore della $T$ target. In questo modo si rallenta e si controlla maggiormente l'adattamento del gas ai nuovi parametri.
Nei $3.8\cdot10^5$ step successivi il sistema viene lasciato evolvere senza riscalamenti e alla conclusione di questo processo vengono effettuate le misurazioni.

```python
T0=120*np.loadtxt("../../Esercizio4/Risultati/Gas/gas_equilibration_temp.dat")

plt.figure(figsize=(15,4))
plt.title("System equilibration",fontsize=16)
plt.plot(T0,color='#117733',lw=1.5,label='System temperature')
plt.axvspan(xmin=0, xmax=25000, color='grey',alpha=0.15,label='Equilibration phase')
plt.hlines(1.2*120,0,400000,color='black',ls='--',label='Target temperature')
plt.ylabel("Temperature [k]",fontsize=14)
plt.xlabel("N step",fontsize=14)
plt.grid(True)
plt.xlim(-5000,410000)
plt.legend(fontsize=14)
plt.show()
```

Come possiamo vedere nel caso della fase gassosa le fluttuazioni sono decisamente più accentuate che nelle fasi precedenti. Seppur con queste difficoltà la temperatura al variare del numero di step sembra mantenersi intorno al valore target di $144 K$ dopo la fase di equilibrazione.

```python
K=eps*np.loadtxt("../../Esercizio4/Risultati/Gas/gas_ekin.dat")
V=eps*np.loadtxt("../../Esercizio4/Risultati/Gas/gas_epot.dat")
E=eps*np.loadtxt("../../Esercizio4/Risultati/Gas/gas_etot.dat")
T=120*np.loadtxt("../../Esercizio4/Risultati/Gas/gas_temp.dat")

x = np.arange(1,len(K)+1,1)

plt.figure(figsize=(18,12))
plt.title(" ")
plt.subplot(2,2,1)
plt.errorbar(x,V.T[0], yerr=V.T[1], color='#332288', alpha=0.8, label='Potential energy', linewidth=2)
plt.ylabel("Potential energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,2)
plt.errorbar(x,K.T[0], yerr=K.T[1], color='#661100', alpha=0.8, label='Kinetic energy', linewidth=2)
plt.ylabel("Kinetic energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,3)
plt.errorbar(x,E.T[0], yerr=E.T[1], color='#882255', alpha=0.8, label='Total energy', linewidth=2)
plt.ylabel("Total energy [J]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.subplot(2,2,4)
plt.errorbar(x,T.T[0], yerr=T.T[1], color='#117733', alpha=0.8, label='Temperature', linewidth=2)
plt.ylabel("Temperature [K]",fontsize=16)
plt.xlabel("# blocks",fontsize=16)
plt.grid(True)

plt.suptitle('System measurements', fontsize=22, y=0.94)
plt.show()
```

I dati riportati nei quattro grafici sono relativi a una simulazione di $1\cdot10^6$ step suddivisi in $100$ blocchi. Nel caso della fase gassosa l'andamento per tutte le grandezze è ancora più oscillante che nella fase liquida e quindi, meno semplice da equilibrare. Per migliorare il trend si potrebbero usare dei parametri iniziali differenti nella fase di equilibrazione o allungare ulteriormente la taglia dei blocchi in modo tale da "catturare" le fluttuazioni del gas.
