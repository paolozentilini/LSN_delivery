```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 10**6
M = 10**3
L = 10**3
X = np.zeros(N)
Y = np.zeros(N)
Z = np.zeros(N)
r = np.zeros(L)
fileXYZ = open('1s_path_uniform.dat', 'r')

for i in range(N):
    X[i], Y[i], Z[i] = fileXYZ.readline().split()
for j in range(L):
    _sum=0
    for i in range(j*M,(j+1)*M,1):
         _sum += np.sqrt(X[i]*X[i] + Y[i]*Y[i] + Z[i]*Z[i])
    r[j] = _sum/M

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
x_blocks = np.arange(2,101,1)
data_eq = np.loadtxt('1s_equilibration_uniform.dat')

plt.title("Fase di equilibrazione (M = 1e6, N=100).")
plt.plot(x_blocks, data_eq.T[0], color='teal', alpha=0.8, label='Simulazione', linewidth=2)
plt.hlines(1.5,2,100, color='salmon', linestyle='-', label="Valore teorico atteso", linewidth=2, alpha=0.9)
plt.ylabel(r'$<\Psi_{1,0,0}|\quad r \quad|\Psi_{1,0,0}> \,\, [\mathrm{a_0}]$')
plt.xlabel("#blocchi")
plt.legend()
plt.grid(True)
plt.ylim(1.2,5)
plt.show()
```

```python
data0=np.loadtxt("1s_uniform.dat")
data1=np.loadtxt("1s_gaussian.dat")
Nblocks = 100
Nmeasures = 1e6
M = Nmeasures/Nblocks
x=np.arange(2,Nblocks+1,1)
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.title("Uniform distribution (M = 1e6, N=100).")
plt.errorbar(x, data0.T[0], yerr=data0.T[1], color='green', alpha=0.5, label='Simulation', linewidth=2)
plt.hlines(1.5,0,Nblocks, color='red', linestyle='-', label="Theoric expected value", linewidth=2, alpha=0.5)
plt.ylabel(r'$<\Psi_{1,0,0}|\quad r \quad|\Psi_{1,0,0}> \,\, [\mathrm{a_0}]$')
plt.xlabel("#blocks")
plt.legend()
plt.grid(True)
plt.ylim(1.45,1.62)

plt.subplot(1,2,2)
plt.title("Gaussian distribution (M = 1e6, N=100).")
plt.errorbar(x, data1.T[0], yerr=data1.T[1], color='blue', alpha=0.5, label='Simulation', linewidth=2)
plt.hlines(1.5,0,Nblocks, color='red', linestyle='-', label="Theoric expected value", linewidth=2, alpha=0.5)
plt.ylabel(r'$<\Psi_{1,0,0}|\quad r \quad|\Psi_{1,0,0}> \,\, [\mathrm{a_0}]$')
plt.xlabel("#blocks")
plt.legend()
plt.grid(True)
plt.ylim(1.45,1.62)
plt.show()
```

```python
data2=np.loadtxt("2p_uniform.dat")
data3=np.loadtxt("2p_gaussian.dat")

plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.title("Uniform distribution (M = 1e6, N=100).")
plt.errorbar(x, data2.T[0], yerr=data2.T[1], color='green', alpha=0.5, label='Simulation', linewidth=2)
plt.hlines(5,0,1e6, color='red', linestyle='-', label="Theoric expected value", linewidth=2, alpha=0.5)
plt.ylabel(r'$<\Psi_{2,1,0}|\quad r \quad|\Psi_{2,1,0}> \,\, [\mathrm{a_0}]$')
plt.xlabel("#throws")
plt.legend()
plt.grid(True)
plt.ylim(4.9,5.25)

plt.subplot(1,2,2)
plt.title("Gaussian distribution (M = 1e6, N=100).")
plt.errorbar(x, data3.T[0], yerr=data3.T[1], color='blue', alpha=0.5, label='Simulation', linewidth=2)
plt.hlines(5,0,1e6, color='red', linestyle='-', label="Theoric expected value", linewidth=2, alpha=0.5)
plt.ylabel(r'$<\Psi_{2,1,0}|\quad r \quad|\Psi_{2,1,0}> \,\, [\mathrm{a_0}]$')
plt.xlabel("#throws")
plt.legend()
plt.grid(True)
plt.ylim(4.9,5.25)
plt.show()
```
